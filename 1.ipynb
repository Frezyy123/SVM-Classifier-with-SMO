```python
import pandas as pd
import scipy.io

mat = scipy.io.loadmat('/home/alexandr/Dropbox/task.mat')



data = pd.DataFrame(data=mat['data'])
data.columns = ['F1','F2','Answer']
import seaborn as sns
%matplotlib inline

data = data.sample(frac=1).reset_index(drop=True)

# sns.pairplot(data=data,hue='Answer')
from sklearn.preprocessing import MinMaxScaler

scaled_data_np = MinMaxScaler().fit_transform(data[['F1','F2']],data['Answer'])
scaled_data_pd = pd.DataFrame(scaled_data_np)

scaled_data_pd.describe()
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
X = scaled_data_pd
y=data['Answer']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4)

model = SVC(kernel='rbf',C=100)

model.fit(X_train,y_train)


expected = y_test
expected
predicted = model.predict(X_test)
metrics.accuracy_score(expected,predicted)
```
